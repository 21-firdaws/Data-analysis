__Step 8. Did you notice the original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob.__

__Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)__

```python
def majority(age):
    if age>17:
        return True
    else:
        return False
```

```python
stud_alcohol['legal_drinker'] = stud_alcohol['age'].apply(majority)
stud_alcohol.head()
```
